## Q2 (4 points) — Basic handcrafted features  
Create these columns and then display the DataFrame:
- `word_count` = number of words  
- `char_count` = number of characters  
- `avg_word_len` = average word length (ignore punctuation)  
- `excl_count` = number of `!` characters  

Print: `id, word_count, char_count, avg_word_len, excl_count`.


```python
# Write your Answer here (Q2)
import pandas as pd
import string

# Load the data into a DataFrame
df = pd.read_csv('product_reviews.txt', sep='\t')

# 1. Create word_count (number of words)
df['word_count'] = df['text'].apply(lambda x: len(x.split()))

# 2. Create char_count (number of characters)
df['char_count'] = df['text'].apply(len)

# 3. Create avg_word_len (average word length, ignoring punctuation)
def get_avg_word_len(text):
    # Remove punctuation using a translation table
    clean_text = text.translate(str.maketrans('', '', string.punctuation))
    words = clean_text.split()
    if not words:
        return 0.0
    return sum(len(word) for word in words) / len(words)

df['avg_word_len'] = df['text'].apply(get_avg_word_len)

# 4. Create excl_count (number of '!' characters)
df['excl_count'] = df['text'].apply(lambda x: x.count('!'))

# Display the specified columns
print(df[['id', 'word_count', 'char_count', 'avg_word_len', 'excl_count']])
```

## Q3 (6 points) — Bag-of-Words (CountVectorizer)  
Use `CountVectorizer(stop_words="english")` on `df["text"]`. Print:
1) vocabulary size (number of features)  
2) top 10 words by **total count** across all documents (word + count)


```python
# # Write your Answer here (Q3)

from sklearn.feature_extraction.text import CountVectorizer

# 1. Initialize CountVectorizer with English stop words
vectorizer = CountVectorizer(stop_words="english")

# 2. Fit and transform the text data
X = vectorizer.fit_transform(df["text"])

# 3. Get the vocabulary and count total occurrences
vocab = vectorizer.get_feature_names_out()
word_counts = X.toarray().sum(axis=0)

# 4. Extract results
vocab_size = len(vocab)
word_freq = list(zip(vocab, word_counts))
top_10_words = sorted(word_freq, key=lambda x: x[1], reverse=True)[:10]

# Print requested information
print(f"Vocabulary size (number of features): {vocab_size}")
print("\nTop 10 words by total count across all documents:")
for word, count in top_10_words:
    print(f"{word}: {count}")
```

## Q4 (4 points) — Bigram features  
Use `CountVectorizer(stop_words="english", ngram_range=(2,2))`.  
Print the top 5 bigrams by total count (bigram + count).


```python
# Write your Answer here (Q4)
from sklearn.feature_extraction.text import CountVectorizer

# 1. Initialize CountVectorizer specifically for bigrams (2-word phrases)
# stop_words="english" filters out filler words like 'the' or 'is'
# ngram_range=(2,2) ensures we only look for pairs of words
vectorizer = CountVectorizer(stop_words="english", ngram_range=(2,2))

# 2. Fit and transform the text data from the DataFrame
X = vectorizer.fit_transform(df["text"])

# 3. Get the list of bigrams and their total counts across all reviews
vocab = vectorizer.get_feature_names_out()
bigram_counts = X.toarray().sum(axis=0)

# 4. Sort and display the top 5 bigrams
bigram_freq = sorted(list(zip(vocab, bigram_counts)), key=lambda x: x[1], reverse=True)

print("Top 5 bigrams by total count:")
for bigram, count in bigram_freq[:5]:
    print(f"{bigram}: {count}")
```

## Q5 (4 points) — TF-IDF features  
Use `TfidfVectorizer(stop_words="english", ngram_range=(1,2))`.  
Compute the **average TF-IDF** score of each term across documents and print the top 5 terms (term + avg score).


```python
# Write your Answer here  (Q5)
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Load the data
filename = 'product_reviews.txt'
df = pd.read_csv(filename, sep='\t')

# 1. Initialize TfidfVectorizer with the specified parameters
vectorizer = TfidfVectorizer(stop_words='english', ngram_range=(1,2))

# 2. Fit and transform the text column
tfidf_matrix = vectorizer.fit_transform(df['text'])

# 3. Get all the feature names (words and bigrams)
feature_names = vectorizer.get_feature_names_out()

# 4. Compute the average TF-IDF score for each term across all documents
# We calculate the mean of each column (feature) in the sparse matrix
avg_tfidf_scores = tfidf_matrix.mean(axis=0).A1  # .A1 flattens the result into a simple 1D array

# 5. Pair each term with its average score and sort by score (descending)
term_scores = list(zip(feature_names, avg_tfidf_scores))
top_5_tfidf = sorted(term_scores, key=lambda x: x[1], reverse=True)[:5]

# 6. Print the top 5 terms and their scores
print("Top 5 terms by average TF-IDF score:")
for term, score in top_5_tfidf:
    print(f"{term}: {score:.4f}")
```
