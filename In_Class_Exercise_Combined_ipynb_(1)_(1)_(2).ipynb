## Reflection (optional, no extra points)
Briefly explain the difference between:
- sentiment analysis
- text classification
- prompt engineering

1. Sentiment Analysis
This is a specific type of text analysis used to determine the emotional tone behind a body of text.

Goal: To identify whether a piece of writing is positive, negative, or neutral (and sometimes specific emotions like angry, happy, or frustrated).

Use Case: Analyzing customer reviews, monitoring social media brand mentions, or gauging public opinion on a topic.

2. Text Classification
This is a broader "umbrella" category that involves assigning predefined labels to a piece of text based on its content.

Goal: To organize text into organized groups.

Relationship: Sentiment Analysis is actually a form of Text Classification (where the labels are "Positive," "Negative," etc.).

Other Use Cases: Spam detection (labeling email as "Spam" or "Not Spam"), topic labeling (labeling news as "Sports," "Politics," or "Tech"), and intent recognition in chatbots.

3. Prompt Engineering
Unlike the first two, which are tasks or outcomes, this is a technique or method for interacting with Generative AI (like LLMs).

Goal: To craft, refine, and optimize "prompts" (inputs) to get the most accurate or high-quality response from an AI model.

Key Difference: While you can use Prompt Engineering to perform sentiment analysis or text classification (by telling an AI "Classify this text..."), it is the process of communication with the model rather than the specific task of labeling data.
